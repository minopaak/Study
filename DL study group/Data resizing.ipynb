```python
import os
import shutil
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from PIL import Image
```

```python
# 데이터 한 폴더에 모으기

from re import A
image_dir = "/content/M_3(2020.09.22)" # Img_dir
file = os.listdir(image_dir)

if not os.path.exists("/content/M_3"):
  os.mkdir("/content/M_3")

for i in range(len(file)):
  path = image_dir
  file2 = os.listdir(path)
  for a in file2:
    if(a[17:] == "NgImage"):
      path2 = path +"/"+ a
      file3 = os.listdir(path2)
      for c in file3:
        shutil.copy(path2+"/"+c, "/content/M_3")
```

```python
# 이미지 보기
im = Image.open("/content/M_3/001_20200622_203305(8).bmp")
plt.imshow(im)
plt.show()
```

```python
# 이미지 사이즈 확인
print(im.size)
```

```python
# 이미지 리사이징 테스트
im = Image.open("/content/M_3/001_20200622_203305(8).bmp")
resized_im = im.resize((576, 576)) # replace the tuple with your desired size
plt.imshow(resized_im)
plt.show()
```

```python
# 전체 이미지 리사이징

image_dir = "/content/M_3"

for filename  in os.listdir(image_dir):
  if filename.startswith("001"):
    im = Image.open(os.path.join(image_dir, filename))
    resized_im = im.resize((576, 576))
    resized_im.save(os.path.join(image_dir, filename))
```
