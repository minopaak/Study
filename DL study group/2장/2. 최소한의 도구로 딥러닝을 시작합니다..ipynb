###넘파이를 소개합니다.

넘파이(Numpy)는 파이썬의 핵심 과학 패키지 중 하나이다. 많은 머신러닝과 딥러닝 패키지가 넘파이를 기반으로 구현되었고 넘파이 인터페이스를 계승하고 있기 때문에 넘파이는
꼭 익혀야 하는 파이썬 패키지 중 하나이다.

###넘파이 준비하기

다차원 배열은 파이썬 리스트로 만들어 사용해도 되지만 파이썬 리스트로 만든 배열은 배열의 크기가 커질수록 성능이 떨어진다는 단점이 있다.
이런경우에 바로 넘파이를 사용한다. 넘파이는 저수준 언어로 다차원 배열을 구현했기 때문에 배열의 크기가 커져도 높은 성능을 보장한다. 또 배열을 이용한 다양한 통계, 수학 함수도 제공한다.

1. 코랩에서 넘파이 임포트하고 버전 확인하기

```python
import numpy as np
print(np.__version__)
```

### 넘파이로 배열 만들기

넘파이를 임포트했으니 넘파이를 이용하여 배열을 하나 만들어보자.
알아야 할 점은 넘파이는 파이썬의 리스트처럼 숫자와 문자열을 함께담을 수 없다.


1. array() 함수로 2차원 배열 만들기.

```python
my_arr = np.array([[10,20,30],[40,50,60]])
print(my_arr)
```

2. type() 함수로 넘파이 배열인지 확인하기

array() 함수는 파이썬 리스트를 입력 받아 넘파이 배열을 만든다. 그런데 출력된 결과를 보면 파이썬 리스트와 형태가 비슷하다. my_arr에 저장된 값이 파이썬 리스트가 아니라 넘파이 배열임을 확인해 보자. type() 함수로 객체의 타입을 확인해보자.

```python
type(my_arr)
```

결과를 보면 my_arr가 넘파이 배열 클래스인 ndarray의 객체임을 알 수 있다. 그런데 이번에 입력한 코드를 보면 print() 함수를 사용하지 않았는데도 결과가 출력되었다. 코드셀의 마지막 줄은 print() 함수를 사용하지 않아도 자동으로 그 결과가 출력된다.

3. 넘파이 배열에서 요소 선택하기

넘파이 배열의 요소를 선택하는 방법은 파이썬 리스트에서 요소를 선택하는 방법과 동일하다. 인덱스는 0부터 시작하므로 첫 번째 행의 세 번째 요소를 선택하려면 다음과 같이 사용한다.

```python
my_arr[0][2]
```

4. 넘파이 내장 함수 사용하기

넘파이 소개를 마치기 전에 넘파이가 제공하는 편리한 내장 함수 하나만 살펴보자. 파이썬 리스트의 요소를 모두 더하려면 for문을 만들어한다. 넘파이에서는 sum()함수를 이용해 배열의 요소를 모두 더할 수 있다. for문을 사용하는 것보다 실행속도도 빠르고 코드도 간결하다.

```python
np.sum(my_arr)
```

###맷플롯립으로 그래프 그리기

맷플롯립(Matplotlib)은 파이썬 과학 생태계의 표준 그래프 패키지이다.
다른 좋은 그래프 도구들도 있지만 맷플롯립에서 제공하는 기능으로 대부분의 그래프를 그릴 수 있다.

맷플롯립도 관례에 따라 별칭 plt를 사용한다.

```python
import matplotlib.pyplot as plt
```

여기서 선 그래프와 산점도를 그려보자.

1. 선 그래프 그리기

선 그래프를 그리려면 x축의 값과 y축의 값을 맷플롯립의 plot() 함수에 전달해야 한다. 그런 다음 show() 함수를 이용하여 화면에 그래프를 출력하면 된다. 다음은 y = $x^2$ 그래프를 그린 것이다.

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

2. 산점도 그리기

자주 사용하는 또 하나의 그래프는 산점도 (scatter plot)이다. 산점도는 데이터의 x축, y축 값의 이용하여 점으로 그래프를 그린 것이다. 이때 show() 함수를 사용하지 않아도 맷플롯립이 자동으로 그래프를 그려주지만 이 함수를 사용하면 조금 더 깔끔하고 보기 좋은 그래프가 그려진다.

```python
plt.scatter([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

3. 넘파이 배열로 산점도 그리기

이번에는 넘파이 배열을 사용하여 산점도를 그려보자. 파이썬의 과학 패키지들은 예외 없이 넘파이 배열로 데이터를 주고받을 수 있도록 만들어져 있다. 여기에서는 넘파이의 random.randn() 함수를 사용하여 표준 정규 분포를 따르는 난수(random number)를 만들고 그 값을 이용하여 산점도를 그리자.

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y)
plt.show()
```

데이터 분석과 그래프 도구는 뗄래야 뗄 수 없는 사이다. 위와 같은 방법으로는 데이터를 시각화하면 데이터에서 직관을 얻기가 쉽다.x
