### 1 - 1 인공지능을 소개합니다.

인공지능(Artificial Inteligence)은 '사람의 지능을 만들기 위한 시스템이나 프로그램'을 말한다. 예를들어 알파고는 '사람처럼 바둑을 둘 수 있게 만들어진 프로그램'이고 음성 비서는 '사람이 음성으로 질문한 것에 대한 답을 할 수 있도록 만들어진 프로그램이다.' 인공지능을 '사람을 완벽하게 대체할 수도 있는 기술' 이라고 오해하기도 한다. 정말 그럴까?




### 사람 같은 인공지능은 강 인공지능이라고 부른다.

사실 인공지능은 강 인공지능(strong AI)과 약 인공지능(weak AI)으로 나눌 수 있다. 강 인공지능은 '사람과 구분이 안 될 정도로 강한 성능을 가진 인공지능'을 말한다. 여기서 '사람과 구분이 안 될 정도로 강한 성능'이란 영화에 등장하는 인공지능을 떠올리면 쉽게 상상할 수 있다. 그렇다면 현실에서 이런 강 인공지능을 쉽게 만들 수 있을까?

현재는 강 인공지능을 만들 수 있는 방법에 대해 아는 사람이 아무도 없다. 심지어 강 인공지능이 나올 만한 시기도 예측하기 어렵다. 그래서 많은 인공지능 과학자들은 강 인공지능에 대해 섣불리 예측하는 것을 꺼린다. 지금까지 발전을 거듭하고 있는 인공지능 기술은 모두 약 인공지능 기술이다.


### 특정 영역에서 작업을 수행하는 인공지능을 약 인공지능이라고 부른다.

약 인공지능은 특정 영역에서 작업을 수행하는 인공지능을 말한다. 운전 보조, 질문 답변, 검색과 같은 일들을 수행한다. 약 인공지능은 주변에서 쉽게 찾아볼 수 있다.

약 인공지능은 특정 영역에서 아주 뛰어난 성능을 발휘하고 있으며 우리의 생활을 보조하는 기술로 많이 사용되고 있다. 그렇다면 머신러닝, 딥러닝은 어디에 포함될까? 두 기술은 모두 약 인공지능에 포함된다.

###머신러닝과 딥러닝 그리고 인공지능의 한계

머신러닝과 딥러닝은 약 인공지능에 포함되는 기술이고, 딥러닝은 머신러닝에 포함되는 기술이다. 머신러닝은 알고리즘의 종류에 따라 더 세부적으로 나눌 수 있다. 그 중에 인공신경망 알고리즘을 사용하여 만든 것이 딥러닝이다. 또한 인공신경망의 기본 구성 요소는 다른 머신러닝 알고리즘이 발전된 것이다. 딥러닝을 잘 이해하기 위해서는 먼저 머신러닝을 알아야하는 이유가 여기에 있다.

#01 - 2 머신러닝을 소개합니다.

머신러닝(Machine Learning)을 풀어 쓰면 '기계 학습'이다. 기계가 학습한다는 의미는 정확히 무엇일까?

### 머신러닝은 스스로 규칙을 수정한다

먼저 알아야 할 개념은 학습이다. 머신러닝과 딥러닝에서 말하는 학습은 데이터의 규칙을 컴퓨터 스스로 찾아내는 것을 말한다. 전통적인 프로그램은 사람이 규칙을 정하여 프로그래밍하고, 사람이 프로그램의 실행 결과를 보며 규칙을 조금씩 수정한다. 반면에 머신러닝은 사람이 만든 프로그램이기는 하지만 규칙을 스스로 찾아 수정한다. 바로 여기서 말하는 규칙을 찾아 수정하는 과정을 학습 또는 훈련이라고 한다. '규칙'이라는 새로운 개념이 등장했다. 이 개념은 무척 중요하지만 애써 이해하려고 할 필요는 없다. 먼저 '학습'이 무엇인지 구체적으로 알아보고 이어서 규칙을 배워보자.



###머신러닝의 학습 방식을 이해하자.

머신러닝은 학습 방식에 따라 지도 학습(supervised learning), 비지도 학습(unsupervised learning), 강화 확습(reinforcement learning)으로 분류한다. 머신러닝으로 다루는 많은 작업들이 지도 학습에 속한다.

**지도 학습은 입력과 타깃으로 모델을 훈련시킨다.**

만약 전날의 습도에 따라 다음 날에 비과 왔는지 안 왔는지 꾸준히 기록했다면 오늘의 습도를 알아도 그 다음 날 비가 올지 안 올지 예측할 수 있을 것이다. 바로 이것이 지도학습에 속한다. 다음은 전날 습도에 따라 오늘 비가 왔는지 안 왔는지를 동그라미와 엑스로 기록한 데이터를 바탕으로 모델을 지도 학습 방식으로 훈련시킨 모습이다. 이 모델은 훈련을 통해 '오늘 습도가 0.6 이상이면 내일 비가 온다'라는 규칙을 얻었다.

모델을 훈련시키기 위해 사용하는 데이터를 통틀어 '훈련 데이터'라고 부른다. 그리고 훈련 데이터는 '입력'과 '타깃(target)으로 구성되어 있다. 입력은 모델이 풀어야 할 일종의 문제와 같은 것이며, 타깃은 모델이 맞춰야할 정답과 같은 것이다. 문제에 대한 답을 주는 방법으로 모델을 훈련시키는 것이다. 학습을 통해 만들어진 프로그램은 모델(Model)이라 하는데, 모델은 새로운 입력에 대한 예측을 만든다.

지도 학습은 기존의 데이터를 통해 모델을 학습시키고 학습시킨 모델로 새로운 입력에 대한 예측을 할 수 있으므로 내일의 날씨를 예측하거나 스팸 이메일을 분류하는 등의 일을 해결할 때 많이 사용한다. 그런데 지도학습에 가장 어려운 것은 모델을 훈련시키는 것이 아니라 훈련 데이터를 만드는 것이다. 잘못된 입력과 타깃을 훈련 데이터에 포함시키면 잘못된 모델이 만들어질 수 도 있고 데이터의 개수가 너무 적으면 모델을 충분히 훈련시킬 수 없기 때문이다.

<br>
**비지도 학습은 타깃이 없는 데이터를 사용한다.**

타깃이 없는 훈련 데이터를 사용하면 비지도 학습이라고 한다. 기업이 고객의 소비 성향에 따라 그룹을 지정하는 상황을 생각하면 비지도 학습의 개념을 쉽게 이해할 수 있을 것이다. 그룹을 만들기 전까지는 어떤 그룹이 존재하는지 또 몇 개의 그룹이 만들어질지 알 수 없다. 즉, 타깃이 없는 셈이다.비지도 학습의 대표적 예는 군집(clusering)이다. 비지도 학습은 훈련 데이터에 타깃이 없으므로 모델의 훈련 결과를 평가하기 어렵다는 특징이 있다.


<br>
**강화 학습은 주어진 환경으로 부터 피드백을 받아 훈련하다.**

강화 학습은 머신러닝 알고리즘으로 에이전트라는 것을 훈련한다. 훈련된 에이전트는 특정 환경에 최적화된 행동을 수행하고 수행에 대한 '보상과' '현재 상태'를 받는다. 에이전트의 목표는 '최대한 많은 보상을 받는 것'이다. 그래서 에이전트는 주어진 환경에서 아주 많은 행동을 수행하여 학습한다. 강화 학습의 대표적인 알고리즘은 Q-러닝(Q-learnig), SARSA, 인공신경망을 사용한 DQN(Deep Q Network) 등이 있다. 강화 학습의 예로는 딥마인드(DeepMind)의 알파고(AlphaGo)와 같은 게임이나 온라인 광고 등이 있다.

###규칙이란 가중치와 절편을 말한다.

앞에서 머신러닝은 스스로 규칙을 찾는다고 설명을 했다. 그러면 규칙은 정확히 무엇일까? 앞의 지도 학습에서 소개한 '오늘 습도에 따라 내일 비가 올지 안 올지 쳬측하는 모델'의 경우 '오늘 습도가 0.6이상이면 내일 비가 온다.'라는 규칙을 정했다.

대부분의 머신러닝 알고리즘은 훈련 데이터와 규칙의 관계를 식으로 표현할 수 있다.

1.5 * x + 0.1 = y(y가 1 이상이면 다음 날 비가 온다고 예측)

위 식에서 x는 입력이고 y는 타깃이다. 1.5와 0.1이 바로 규칙이다. 이 규칙은 각각 가중치와 절편이라고 부르는데 입력과 곱하는 수가 가중치, 더하는 수가 절편이다. 여기서는  가중치와 절편이 각각 1개인 간단한 식을 규칙의 예를 들었지만, 딥러닝은 이것보다 훨씬 많은 개수의 가중치와 절편을 가진다.

###모델은 머신러닝의 수학적 표현이다

앞에서 만든 수학식이 바로 모델이다. 엄밀히 정의하면 모델은 '훈련 데이터로 학습된 머신러닝 알고리즘'을 말한다/ 그리고 가중치와 절편을 합쳐 모델 파라미터(model parameter)라고 부른다. 앞으로 이런 모델들을 파이썬으로 직접 만들어볼 것이다.

###손실 함수로 모델의 규칙을 수정한다.

앞에서 만든 모델은 습도에 따라 내일 비가 올지 잘 예측하는 것 같다. 그렇다면 훈련 데이터에 새로운 입력과 출력이 있으면 어떨까?

이전에 만든 모델이 새로운 입력을 넣어 예측한 값이 실제 타깃값과 맞지 않는 경우, 이런 경우에 무델의 규칙을 수정해야 한다. 이때 모델의 규칙을 수정하는 기준이 되는 함수를 '손실 함수(loss function)'라 부른다.

**최적화 알고리즘으로 손실 함수의 최솟값을 찾는다.**

손실 함수는 단순히 예측한 갑솨 타깃의 차이를 정의하는 것이므로 손실 함수의 최솟값을 효율적으로 찾는 방법도 필요하다. 이때 최솟값을 효율적으로 찾는 방법을 최적화 알고리즘이라고 부른다. 지금까지 머신러닝을 이해하기 위해 필요한 개념들을 하나씩 짚어보았다.

# 01 - 3 딥러닝을 소개합니다,

### 딥러닝은 인공신경망으로 구성된다.

딥러닝이라는 용어는 인공신경망을 여러 겹으로 쌓은 모습에서 유래되었다. 입력을 받아 출력으로 전달하는 인공신경망을 보며 뇌의 뉴런과 비슷한 점이 많다. 뉴런은 자극을 받아들여 다음 뉴런으로 자극을 전달하는 역할을 수행하기 때문이다. 인공신경망을 여러 겹 쌓으면, 이것이 바로 딥러닝이다.

**딥러닝은 사람의 뇌와 많이 다르다**

인공신경망이라는 이름 때문에 '딥러닝은 사람의 뇌와 연관이 있는 기술이다.라는 오해를 할 수 있다. 물론 뇌의 뉴런에서 영감을 받아 만든 것이다. 하지만 딥러닝은 실제 뇌가 작동한느 방식과는 많이 다르니 오해하지 말자.

###딥러닝은 머신러닝이 처리하기 어려운 데이터를 더 잘 처리한다.

머신러닝의 한 종류인 딥러닝은 머신러닝과 구체적으로 어떤 점이 다를"까? 대표적으로 둘의 차이점을 이야기 할 때 '처리하는 데이터'를 예로 드는 경우가 많다. 연구자들은 머신러닝이 처리하기 어려운 데이터를 딥러닝이 더 잘 처리한다는 것을 알았다.

딥러닝은 보통 '인지'와 관련된 문제를 잘 해결한다고 알려져 있다. 예를 들어 딥러닝은 이미지, 음성, 텍스트에 관련된 분야의 문제를 잘 해결한다. 반면에 머신러닝은 암환자의 검진 데이터나 회사의 매출 데이터와 같은 정형화된 데이터에 대한 문제를 잘 해결한다.
